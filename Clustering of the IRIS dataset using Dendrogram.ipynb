#### The IRIS data consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). The dataset describes four features from each sample: length and width of sepals and petals.

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris
```

## Load the Data

```python
iris = load_iris()
```

```python
iris.feature_names

# The data has four features: 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'
```

```python
# Define features
X = iris.data
```

## Select the Linkage method

The linkage method is a technique used in hierarchical clustering to determine the distance between clusters. It defines the distance between two clusters as a function of the pairwise distances between their members.

There are several linkage methods available, including:

1. Single linkage: The distance between two clusters is defined as the minimum distance between any two points in the two clusters.
2. Complete linkage: The distance between two clusters is defined as the maximum distance between any two points in the two clusters.
3. Average linkage: The distance between two clusters is defined as the average distance between all pairs of points in the two clusters.
3. Ward linkage: The distance between two clusters is defined as the increase in the sum of squared distances when the two clusters are merged.

Different linkage methods can produce different clustering results, and the choice of linkage method can depend on the specific characteristics of the data and the goals of the analysis. For example, complete linkage tends to produce compact, spherical clusters, while single linkage can produce elongated, chain-like clusters. Ward linkage is often preferred when the goal is to minimize the variance within clusters.

```python
# Use Ward linkage method to perform hierarchical clustering
linkage = linkage(X, method= 'ward')
```

## Plot the dendrogram

```python
plt.figure(figsize=(40, 16))
dendrogram(linkage, leaf_rotation=90., leaf_font_size=8. , labels=iris.target,
           color_threshold=15, above_threshold_color='gray')
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel('Iris')
plt.ylabel('Distance')
plt.show()
```
